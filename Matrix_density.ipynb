```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.colors import LogNorm
import numpy as np

# x, y, z, a = np.loadtxt('data/training.csv', delimiter=",")
df = pd.read_csv('data/training.csv')
```

```python
x = np.array( df.user)
```

```python
y = np.array( df.movie)
z = np.array( df.rating)
```

```python
colors = 'mrygb'

map = []
for r in z:
    map.append(colors[r-1])
len(map)
```

```python

f, ax = plt.subplots(figsize=(15,15))
plt.scatter(x,y, marker='.', alpha=.2 , c=map, s=.5)
```

```python

f, ax = plt.subplots(figsize=(15,15))
plt.scatter(x,y, marker='.', alpha=.2 , c=map, s=.1)
```

```python
N = int(len()**.5)
z = z.reshape(N, N)
plt.imshow(z+10, extent=(np.amin(x), np.amax(x), np.amin(y), np.amax(y)),
        cmap=cm.hot, norm=LogNorm())
plt.colorbar()
plt.show()
```
